# Hill Climbing

Hill climbing is a simple optimization algorithm , which aim to maximize the given function - often picked as a first choice for solving new problems due to its simplicity and efficiency. It also has many simple variations and provides a good benchmark for more complicated algorithms.

The idea is very simple - to get to the highest point in a landscape, you take a step in the direction of steepest ascend and repeat until there is no direction which lead you higher. This ia local search strategy and it may or may not work well depending on the shape of the hill, the number of local maxima and the initial position. The hill climbing algorithms can be easilty adapted to serve as Hill Descending Algorithms, which should minimize the function instead.

```python
import benchmarks
import matplotlib.pyplot as plt
```

### Steepest Ascent Hill Climbing

Steepest Ascent proceeds conservatively, changing only one solution feature at a time, and replacing the current best solution with the best one-feature change.

```python
NUM_GENERATIONS = 1000
NUM_FEATURES = 20
x = np.random.uniform(-5, 5, NUM_FEATURES)

generation = 0
results = []
while generation < NUM_GENERATIONS:
    fitness = benchmarks.ackley_function(x)
    results.append(fitness)
    
    offsprings = []
    offsprings.append((x, fitness))
    for i in range(NUM_FEATURES):
        offspring = x.copy()
        
        if np.random.random() < .5:
            offspring[i] = offspring[i] + np.random.uniform()
        else: 
            offspring[i] = offspring[i] - np.random.uniform()
            
        offspring_fitness = benchmarks.ackley_function(offspring)
        offsprings.append((offspring, offspring_fitness))
        
    best_offspring = min(offsprings, key=lambda t: (t[1], -t[0]))
    if (best_offspring[0] == x).all():
        x = np.random.uniform(-1, 1, NUM_FEATURES)
    else:
        x = best_offspring[0]
        
    generation += 1
    
print('With fitness of {}, best solution is:'.format(best_offspring[1]))
print(x)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(results)

plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Steepest Ascent')

plt.show()
```

### Simple Hill Climbing

Like steepest ascent, simple hill climbing also changes one solution feature at a time, but it does it in a more greedy way - as soon as a better solution is found, the current solution is replaced.

```python
NUM_GENERATIONS = 1000
NUM_FEATURES = 20
x = np.random.uniform(-5, 5, NUM_FEATURES)

generation = 0
results = []
while generation < NUM_GENERATIONS:
    parent_fitness = benchmarks.ackley_function(x)
    results.append(parent_fitness)
    replace_flag = False
    
    for i in range(NUM_FEATURES):
        offspring = x.copy()
        
        if np.random.random() < .5:
            offspring[i] = offspring[i] + np.random.uniform()
        else: 
            offspring[i] = offspring[i] - np.random.uniform()
            
        offspring_fitness = benchmarks.ackley_function(offspring)
        if offspring_fitness < parent_fitness:
            x = offspring
            replace_flag = True
    
    if not replace_flag:
        x = np.random.uniform(-1, 1, NUM_FEATURES)
        
    generation += 1
    
print('With fitness of {}, best solution is:'.format(benchmarks.ackley_function(x)))
print(x)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(results)

plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Simple Hill Climbing')

plt.show()
```

### Random Mutation Hill Climbing

Random Mutation fall under the category of stochastic hill climbings, which randomly selects which solution feature to mutate. It can be exactly the same as steepest ascent algorith, with the exception for that fact that the mutated solution is chosen at random.

```python
NUM_GENERATIONS = 1000
NUM_FEATURES = 20
x = np.random.uniform(-5, 5, NUM_FEATURES)

generation = 0
results = []
while generation < NUM_GENERATIONS:
    parent_fitness = benchmarks.ackley_function(x)
    results.append(parent_fitness)
    q = np.random.randint(NUM_FEATURES)
    
    offspring = x.copy()
    if np.random.random() < .5:
        offspring[q] = offspring[q] + np.random.uniform()
    else: 
        offspring[q] = offspring[q] - np.random.uniform()

    offspring_fitness = benchmarks.ackley_function(offspring)
    if offspring_fitness < parent_fitness:
        x = offspring.copy()
        
    generation += 1
    
print('With fitness of {}, best solution is:'.format(benchmarks.ackley_function(x)))
print(x)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(results)

plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Random Mutation Hill Climbing')

plt.show()
```

### Adaptive Hill Climbing

Adaptive Hill Climbing is very similar to Random Mutation and also fall under the category of stochastic hill climbing algorithms. The difference is that every solution feature is mutated with some probability before the mutated solution is compared with the current best solution.

```python
NUM_GENERATIONS = 1000
NUM_FEATURES = 20
MUTATION_PROBABILITY = 0.1
x = np.random.uniform(-5, 5, NUM_FEATURES)

generation = 0
results = []
while generation < NUM_GENERATIONS:
    parent_fitness = benchmarks.ackley_function(x)
    results.append(parent_fitness)
    
    offspring = x.copy()
    for i in range(NUM_FEATURES):
        if np.random.random() < MUTATION_PROBABILITY:
            if np.random.random() < .5:
                offspring[i] = offspring[i] + np.random.uniform()
            else: 
                offspring[i] = offspring[i] - np.random.uniform()

    offspring_fitness = benchmarks.ackley_function(offspring)
    if offspring_fitness < parent_fitness:
        x = offspring.copy()
        
    generation += 1
    
print('With fitness of {}, best solution is:'.format(benchmarks.ackley_function(x)))
print(x)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(results)

plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Adaptive Hill Climbing')

plt.show()
```
