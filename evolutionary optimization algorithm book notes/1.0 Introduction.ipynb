# Continuous Optimization

Problems where independent variables can vary continuously.

### Unconstrained Optimization:

When we try to minimize a function, we call the function value the cost function.
When we try to maximize a function, we call the function value the fitness.

$$\min_x f(x) \implies f(x) \text{ is called 'cost' or 'objective'}$$
$$\max_x f(x) \implies f(x) \text{ is called 'fitness' or 'objective'}$$

```python
import math
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import seaborn as sns
```

Consider the following function:

$$f(x) = x^4 + 5x^3 + 4x^2 - 4x + 1$$

f(x) function above is a *quartic polynomial*, also called *fourth order* or *fourth degree*, meaning it has at most three stationary points - or in other words - three values of *x* at which its derivative f'(x) = 0. The derivative of this functions is:

$$f'(x) = 4x^3 + 15x^2 + 4$$

As we can confirm that it equals 0 at the visible spots, where $x=-2.96, x=-1.1 \text{ or } x=0.31$

We can go further and find the second derivative of f(x) at these 3 points:

$$f''(x) = 12x^2 + 30x + 8 = \begin{cases}
    24.33, & \quad x=-2.96 \\
    -10.48, & \quad x=-1.10 \\
    18.45, & \quad x=0.31
    \end{cases}
$$

Second derivative of a function at a *local minimum* is **positive** and the second derivative of a function at a *local maximum* is **negative**. We can confirm that from a graph, where value at -1.10 is a local maximum (negative in 2nd derivative) and other two values at at local minimum (positive in the 2nd derivative).

```python
def f(x):
    return x**4 + 5*x**3 + 4*x**2 - 4*x + 1

x = [x for x in np.linspace(-4, 1, 40)]
y = [f(_x) for _x in x]

sns.lineplot(x, y)
plt.show()
```

Function above has 2 local minima and one global minimum - note that global minimum is also considered a local minimum. Thus, for some function $min_{x} f(x)$ occurs at more than one value of x, meaning that such functions have multiple global minima, which is dependend of a distance metric (epsilon) we define:

$$f(x^*) < f(x) \text{ for all x such that } ||x-x^*|| < \epsilon$$

A global minimum can be defined as:

$$f(x^*) \leq f(x) \text{ for all x }$$

### Constrained Optimization

A kind of problem where we need to minimize some function f(x) with restrictions on the allowable values of x. Assuming we use the same function as before, but with restriction that $x \geq -1.5$. Based on that, we can see that only one value, $x=0.31$ is the minimizing value of x for the contrained minimization problem.

# Multi-objective Optimization

Most of the problems are both contrained and multi-objective - meaning we are interested in minimizing more than one measure simultaneously. One of the examples of multi-objective problems can look like:

$$\min_x[f(x) \text{ and } g(x)] \quad \text{where } \quad  f(x) = x^4 + 5x^3 + 4x^2 - 4x +1 \\
\quad \quad \quad \text{and } \quad  g(x) = 2(x+1)^2$$

```python
def f(x):
    return x**4 + 5*x**3 + 4*x**2 - 4*x + 1
def g(x):
    return 2*(x+1)**2

x_1 = [x for x in np.linspace(-4, 1, 40)]
y_1 = [f(_x) for _x in x_1]

x_2 = [x for x in np.linspace(-4, 1, 40)]
y_2 = [g(_x) for _x in x_2]

sns.lineplot(x_1, y_1, color='r')
sns.lineplot(x_2, y_2, color='g')
plt.legend(['f(x)', 'g(x)'])
plt.show()
```

As we can see, each function have their global minima in different locations, causing conflicts between 2 objectives. We can construct *Pareto set* and *Pareto Front* by knowing the outcome for each functions given value *x*. Parento set gives as best possible x values for the given solution - however we cannot say anything more about the optimal value of x per se. Insead, that is a question of engineering judgement to select a point along the Pareto from as the ultimate solution. In our example, Parto set and front look as follows:

$$\text{Pareto set: } \quad x^*={x: x \in [-2.96, -2] \text{ or } x \in [-1, 0]} \\
\text{Pareto front: } \quad {(f(x),g(x)): x \in x^*}$$ 

Which is kind of at the places where both graphs overlap - but depending on the problem, best optimization might not necesarrily be at the overlap point.

### Multimodal Optimization

Multimodal optimization problems are problems that have more than one local minimum. We've already seen multimodal problem in unconstrained optimization paragraph, but that one had only 2 local minima, so it was easy to handle - normally such problems have many of those, and it's challanging to discover which minimum is the global minimum. One such example can be Auckley function, which has plenty of local minima even in 2d space, not to mention 20- or 30-dimensions. We could solve this problem by taking the derivative of f(x, y) with respect to x and y, and then solving for $f_{x}(x,y) = f_{y}(x,y) = 0$ in order to find local minima, however, that solving those equations simultanously could be difficult. In those cases, EA can get us much better results.

```python
def auckley(x, y):
    sum_1 = np.power(x, 2) + np.power(y, 2)
    term_1 = -20*np.power(np.e, (-0.2*np.sqrt(sum_1/2)))
    
    sum_2 = np.cos(2*np.pi*x) + np.cos(2*np.pi*y)
    term_2 = -np.power(np.e, sum_2/2)
    
    return 20 + np.e + term_1 + term_2
```

```python
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)

X, Y = np.meshgrid(x, y)
Z = auckley(X, Y)
```

```python
fig = plt.figure(figsize=(15, 10))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='inferno')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Auckley Function Plot')

plt.show()
```

# Combinatorial Optimization

Problems, where independent variables are restricted to a set of discrete values. Travelling Salesman Problem (TSP) is one of the example of combinatorial optimization problems - say, salesman wants to visit 4 cities, starting and ending on his own. The problem aims to answer the questions: in what order should the cities be visited, so that the distance travelled is minimized?

To find the answer to TSP one can do it via *brute-force* (or *exhaustive search*), meaning that all possible combinations of all cities will be checked and compared. It's relatively easy when we only have a small number of cities to consider - the problem becomes much harder, when the number of cities grows. In fact, for *n* cities there are (n-1)! solutions, making it extremely resource consuming problem and using brute force to solve it is definitely not the most efficient solution.
